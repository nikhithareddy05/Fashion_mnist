```python
data = keras.datasets.fashion_mnist # load the fashion mnist data
```

```python
(train_images,train_labels),(test_images,test_labels) = data.load_data() # spliting the data
```

```python
class_names = ["T-Shirt/Top","Trouser","Pullover","Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankle boot"] # assigning labels withs names
```

```python
train_images = train_images/255.0   # normalizing the data
```

```python
train_images[7]
```

```python
test_images = test_images/255.0  # normalizing the daa actual values from 0 to 255
```

```python
model = keras.Sequential([keras.layers.Flatten(input_shape=(28,28)),
                          keras.layers.Dense(120,activation = "relu"),
                         keras.layers.Dense(10,activation = "softmax")])  # tensorflow graph
                        
```

```python
model.compile(optimizer='adam',  
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']) # optimizing the weights and biases
```

```python
model.fit(train_images,train_labels,epochs = 5)  # train the models
```

```python
test_loss,test_acc = model.evaluate(test_images,test_labels)   # evalulate the model
```

```python
print("Tested_Accuracy:", test_acc)
```

```python
prediction = model.predict(test_images)  # predict the model
```
